# Spin Parameter Equation

Using the formula derived by Bardeen (1970) for the evolution of the spin parameter:

$$ a^f = \frac{1}{3}r_{lso}^{1/2}\frac{M_{bh}}{M^f_{bh}}\left[ 4 - \left\{ 3 r_{lso}\left( \frac{M_{bh}}{M^f_{bh}} \right)^2 - 2 \right\}^{1/2} \right] $$

with 

$$ r_{lso} = 3 + Z_2 \mp \left[ (3-Z_1)(3+Z_1)(2Z_2)^{1/2} \right]$$

where the sign is taken contrary to the sign of the spin parameter. $Z_1$ and $Z_2$ are defined as:

$$ Z_1 = 1 + (1-a^2)\left[ (1+a)^{1/3} + (1-a)^{1/3} \right] $$

$$ Z_2 = (3a^2 + Z_1^2)^{1/2} $$

```python
#Formula for spin parameter evolution
def spin_evolution( MfM0, a ):
    Z1 = 1 + (1-a**2)**(1/3.)*( (1+a)**(1/3.) + (1-a)**(1/3.) )
    Z2 = (3*a**2 + Z1**2)**(0.5)
    Rlso = 3 + Z2 - np.sign(a)*( (3-Z1)*(3+Z1+2*Z2) )**(0.5)
    af = 1/3.0*Rlso**0.5*(1.0/MfM0)*( 4 - (3*Rlso*MfM0**-2-2)**(0.5) )
    try:
        af[MfM0 > Rlso**0.5] = 0.998
    except:
        None
    return af
```

```python
spin_evolution( np.array([1.2,]), np.array([-0.5,]) )
```

```python
MfM0 = np.linspace( 1, 3.5, 100 )
a0 = np.linspace( -1, 1, 100 )
Mg, ag = np.meshgrid( MfM0, a0 )
af = spin_evolution( Mg, ag )
plt.figure(figsize=(8,8))
plt.contour(af[::,::], extent = (1,3.5,-1,1), colors='black')
plt.imshow(af[::-1,::], extent = (1,3.5,-1,1))
plt.xlabel('$M^f_{bh}/M_{bh}$', fontsize=18)
plt.ylabel('$a$', fontsize=18)
plt.title('Evolution of spin parameter')
plt.colorbar(orientation='horizontal')
plt.grid()
```

```python
plt.plot( MfM0, spin_evolution( MfM0, 0.5 ) )
plt.plot( MfM0, spin_evolution( MfM0, -1.0 ) )
```

Checking different mass intervals

```python
plt.figure( figsize = (10,5) )
Mass_intervals = [ 0.1, 0.01, 0.005 ]
Tmax = 2
M0 = 1
Mf = 2
A0 = 0.05
for dm in Mass_intervals:
    m = M0
    a = A0
    Time = []
    A = []
    while m<=Mf:
        Time.append( Tmax*(m-M0)/(Mf-M0) )
        A.append( a )
        m = m + dm
        if np.random.random()<0.5:
            a = spin_evolution( m/(m-dm), a )
        else:
            a = -spin_evolution( m/(m-dm), -a )
    plt.plot( Time, A, 'o-', lw=2, label='$\Delta m = %1.2f $'%(dm) )
plt.legend( loc='lower right' )
plt.ylabel( 'Spin parameter' )
plt.xlabel( 'Time' )
```

```python
#Cavendish constant [SI]
GC	=	6.67384e-11
#Light speed
CLIGHT	=	2.99792458e8
#Proton Mass
PROTONMASS =    1.67262178e-27
#Thomson constant
THOMPSON =      6.65245873e-29
#1e8 Solar units
E8MSUN = 	1.989e38
#Seconds per year
SECPERYEAR =    3.15569e7
#Solar mass
SOLARMASS =     1.989e30


def m_dot_eddignton( Mbh, rad_eff = 0.1 ):
	return (4*np.pi*GC*CLIGHT*PROTONMASS/(rad_eff*CLIGHT**2*THOMPSON))*Mbh*3.15569e16

def QuasarBHmode( Mdot, Mbh ):
    m = Mdot/m_dot_eddignton( Mbh, rad_eff = 0.1 )
    thrs = 0.002*(Mbh*1.989e40/E8MSUN)**2.0
    thrs[thrs>0.1] = 0.1
    return (m>thrs).astype(int)
```

```python
plt.figure( figsize = (18,6) )

t_quasar = test00.t[QuasarBHmode( test00.mbh, test00.mbh_dot ).astype(bool)][-1]

test00.spin_evolution( mode='chaotic', alpha = 0.1, a0 = 0.1 )
plt.subplot(1,2,1)
plt.plot( test00.tms, test00.aps, '-', label='chaotic', lw=2 )
plt.fill_between([0,t_quasar], -0.4, 1.0, alpha=0.1, color='green')
plt.fill_between([t_quasar,test00.tms[-1]], -0.4, 1.0, alpha=0.1, color='blue')
plt.subplot(1,2,2)
plt.semilogy( test00.tms, test00.dms, '-', label='chaotic', lw=2 )
plt.fill_between([0,t_quasar], 1e-10, 1.0, alpha=0.1, color='green')
plt.fill_between([t_quasar,test00.tms[-1]], 1e-10, 1.0, alpha=0.1, color='blue')

test01.spin_evolution( mode='continuous', alpha = 0.1, a0 = 0.1 )
plt.subplot(1,2,1)
plt.plot( test01.tms, test01.aps, '-', label='prolonged', lw=2 )
plt.subplot(1,2,2)
plt.semilogy( test01.tms, test01.dms, '-', label='prolonged', lw=2 )

plt.subplot(1,2,1)
plt.title('Model using $M_{bh}$ from the simulation')
plt.ylabel("Spin parameter")
plt.xlabel("time [Gyr]")
plt.legend(loc='center right')
plt.ylim((-0.4,1))
plt.xlim((0.0,test00.tms[-1]))
plt.subplot(1,2,2)
plt.title('Model using $M_{bh}$ from the simulation')
plt.ylabel("Mass of episode")
plt.xlabel("time [Gyr]")
plt.legend(loc='center right')
plt.xlim((0.0,test00.tms[-1]))
```
