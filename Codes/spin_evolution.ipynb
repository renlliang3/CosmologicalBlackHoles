# Spin Parameter Equation

Using the formula derived by Bardeen (1970) for the evolution of the spin parameter:

$$ a^f = \frac{1}{3}r_{lso}^{1/2}\frac{M_{bh}}{M^f_{bh}}\left[ 4 - \left\{ 3 r_{lso}\left( \frac{M_{bh}}{M^f_{bh}} \right)^2 - 2 \right\}^{1/2} \right] $$

with 

$$ r_{lso} = 3 + Z_2 \mp \left[ (3-Z_1)(3+Z_1)(2Z_2)^{1/2} \right]$$

where the sign is taken contrary to the sign of the spin parameter. $Z_1$ and $Z_2$ are defined as:

$$ Z_1 = 1 + (1-a^2)\left[ (1+a)^{1/3} + (1-a)^{1/3} \right] $$

$$ Z_2 = (3a^2 + Z_1^2)^{1/2} $$

```python
#Formula for spin parameter evolution
def spin_evolution( MfM0, a ):
    Z1 = 1 + (1-a**2)**(1/3.)*( (1+a)**(1/3.) + (1-a)**(1/3.) )
    Z2 = (3*a**2 + Z1**2)**(0.5)
    Rlso = 3 + Z2 - np.sign(a)*( (3-Z1)*(3+Z1+2*Z2) )**(0.5)
    af = 1/3.0*Rlso**0.5*(1.0/MfM0)*( 4 - (3*Rlso*MfM0**-2-2)**(0.5) )
    af[MfM0 > Rlso**0.5] = 0.998
    return af
```

```python
spin_evolution( np.array([1.2,]), np.array([-0.5,]) )
```

```python
MfM0 = np.linspace( 1, 3.5, 100 )
a0 = np.linspace( -1, 1, 100 )
Mg, ag = np.meshgrid( MfM0, a0 )
af = spin_evolution( Mg, ag )
plt.figure(figsize=(8,8))
plt.contour(af[::,::], extent = (1,3.5,-1,1), colors='black')
plt.imshow(af[::-1,::], extent = (1,3.5,-1,1))
plt.xlabel('$M^f_{bh}/M_{bh}$', fontsize=18)
plt.ylabel('$a$', fontsize=18)
plt.title('Evolution of spin parameter')
plt.colorbar(orientation='horizontal')
plt.grid()
```

```python
plt.plot( MfM0, spin_evolution( MfM0, 0.5 ) )
plt.plot( MfM0, spin_evolution( MfM0, -1.0 ) )
```
