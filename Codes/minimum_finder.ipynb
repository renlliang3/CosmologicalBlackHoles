## Taking particles close to the box's center

```python
#Filtering particles closer to the box's center
Center = Coordinates[id_bound]
#Number of selected particles
Npart = 200
#Properties
mask = np.argsort(np.linalg.norm( Coordinates - 1*Center, axis=1 ))[:Npart]
CoordCls = Coordinates[mask]
PotCls = Potential[mask]

#1D projection
fig = plt.figure( figsize = (15,5) )
ax2 = fig.add_subplot( 131 )
ax2.plot( CoordCls[:,[0]], PotCls, '.' )
ax2 = fig.add_subplot( 132 )
ax2.plot( CoordCls[:,[1]], PotCls, '.' )
ax2 = fig.add_subplot( 133 )
ax2.plot( CoordCls[:,[2]], PotCls, '.' )
```
