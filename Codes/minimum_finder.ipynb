## Tracking particles close to the most bounded particle

The closer particles to the most bounded one are taken, and then, the potential is projected in every axis. This way, the multidimensional minimization problem is reduced to 3 one-dimensional ones, which might signify a reduction in the computing time as the parameter space to be explored is drastically reduced.

```python
#Filtering particles closer to the box's center
Center = Coordinates[id_bound]
#Distance from most bounded particle
Distance = 3
#Properties
mask = np.linalg.norm( Coordinates - 1*Center, axis=1 ) <= Distance
CoordCls = Coordinates[mask]
PotCls = Potential[mask]

#1D projection
fig = plt.figure( figsize = (15,5) )
ax1 = fig.add_subplot( 131 )
ax1.plot( CoordCls[:,0], PotCls, '.' )
ax1.set_ylim( (PotCls.min(), PotCls.max()) )
ax1.set_ylabel( "Potential" )
ax1.set_xlabel( "X [kpc]" )

ax2 = fig.add_subplot( 132 )
ax2.plot( CoordCls[:,1], PotCls, '.' )
ax2.set_ylim( (PotCls.min(), PotCls.max()) )
ax2.set_xlabel( "Y [kpc]" )

ax3 = fig.add_subplot( 133 )
ax3.plot( CoordCls[:,2], PotCls, '.' )
ax3.set_ylim( (PotCls.min(), PotCls.max()) )
ax3.set_xlabel( "Z [kpc]" )
```

## Extracting contouring particles in every projection

In order to fit the contour of the potential in every projection using a second order expansion, the contouring particles are extracted. To do so, a grid is set in every axis and the lowest potential particles are taken.

```python
#Number of bins
Nbins = 30
#Finding most bounded particles delimiting potential in every projection
Rps = []
Pps = []
for ip in xrange(3):
    #Building grid
    R = []
    P = []
    for ig in xrange(Nbins):
        coords_shift = CoordCls[:,ip]-(Center[ip]-Distance)
        mask = (2.0*ig*Distance/Nbins <= coords_shift)*(coords_shift < 2.0*(ig+1)*Distance/Nbins)
        try:
            id_min = np.argsort(PotCls[mask])[0]
            R.append( CoordCls[mask][id_min,ip] )
            P.append( PotCls[mask].min() )
        except:
            None
    R = np.array( R )
    P = np.array( P )
    Rps.append( R )
    Pps.append( P )
Rps = np.array( Rps )
Pps = np.array( Pps )
```

```python
#1D projection
fig = plt.figure( figsize = (15,5) )
ax1 = fig.add_subplot( 131 )
ax1.plot( CoordCls[:,0], PotCls, '.' )
ax1.plot( Rps[0], Pps[0], 'o-', lw=1, color='red' )
ax1.set_ylim( (PotCls.min(), PotCls.max()) )
ax1.set_ylabel( "Potential" )
ax1.set_xlabel( "X [kpc]" )

ax2 = fig.add_subplot( 132 )
ax2.plot( CoordCls[:,1], PotCls, '.' )
ax2.plot( Rps[1], Pps[1], 'o-', lw=1, color='red' )
ax2.set_ylim( (PotCls.min(), PotCls.max()) )
ax2.set_xlabel( "Y [kpc]" )

ax3 = fig.add_subplot( 133 )
ax3.plot( CoordCls[:,2], PotCls, '.' )
ax3.plot( Rps[2], Pps[2], 'o-', lw=1, color='red' )
ax3.set_ylim( (PotCls.min(), PotCls.max()) )
ax3.set_xlabel( "Z [kpc]" )
```

## Offsetting deformation caused by Black Hole

As shown above, the interaction of the SMBH with the background particles causes a considerable deformation in the potential which can alter the detection of potential minimum when using the most bounded particle. Actually this was what caused the unrealistic kick of the BH in the circular decaying orbit in the simulation 'hernq_drag_BH_1e6'

```python
#Loading particles with BH correction
CoordinatesC, PotentialC, id_boundC, r_mpC, r_bhC = snapshot_reader( DataFolder, 'hernq_kick_BH_1e6', snap = 4, BH_offset = True )

#Filtering particles closer to the box's center
CenterC = CoordinatesC[id_bound]
#Distance from most bounded particle
Distance = 3
#Properties
mask = np.linalg.norm( CoordinatesC - 1*CenterC, axis=1 ) <= Distance
CoordClsC = CoordinatesC[mask]
PotClsC = PotentialC[mask]

#Finding most bounded particles delimiting potential in every projection
RpsC = []
PpsC = []
for ip in xrange(3):
    #Building grid
    R = []
    P = []
    for ig in xrange(Nbins):
        coords_shift = CoordClsC[:,ip]-(CenterC[ip]-Distance)
        mask = (2.0*ig*Distance/Nbins <= coords_shift)*(coords_shift < 2.0*(ig+1)*Distance/Nbins)
        try:
            id_min = np.argsort(PotClsC[mask])[0]
            R.append( CoordClsC[mask][id_min,ip] )
            P.append( PotClsC[mask].min() )
        except:
            None
    R = np.array( R )
    P = np.array( P )
    RpsC.append( R )
    PpsC.append( P )
RpsC = np.array( RpsC )
PpsC = np.array( PpsC )

#1D projection
fig = plt.figure( figsize = (15,5) )
ax1 = fig.add_subplot( 131 )
ax1.plot( CoordClsC[:,0], PotClsC, '.' )
ax1.plot( RpsC[0], PpsC[0], 'o-', lw=1, color='red' )
ax1.set_ylim( (PotClsC.min(), PotClsC.max()) )
ax1.set_ylabel( "Potential" )
ax1.set_xlabel( "X [kpc]" )

ax2 = fig.add_subplot( 132 )
ax2.plot( CoordClsC[:,1], PotClsC, '.' )
ax2.plot( RpsC[1], PpsC[1], 'o-', lw=1, color='red' )
ax2.set_ylim( (PotClsC.min(), PotClsC.max()) )
ax2.set_xlabel( "Y [kpc]" )

ax3 = fig.add_subplot( 133 )
ax3.plot( CoordClsC[:,2], PotClsC, '.' )
ax3.plot( RpsC[2], PpsC[2], 'o-', lw=1, color='red' )
ax3.set_ylim( (PotClsC.min(), PotClsC.max()) )
ax3.set_xlabel( "Z [kpc]" )
```

## Effects of rotation

```python
#Defining rotation Matrix
def rotation_matrix(axis, theta):
    """
    Return the rotation matrix associated with counterclockwise rotation about
    the given axis by theta radians.
    """
    axis = np.asarray(axis)
    theta = np.asarray(theta)
    axis = axis/math.sqrt(np.dot(axis, axis))
    a = math.cos(theta/2.0)
    b, c, d = -axis*math.sin(theta/2.0)
    aa, bb, cc, dd = a*a, b*b, c*c, d*d
    bc, ad, ac, ab, bd, cd = b*c, a*d, a*c, a*b, b*d, c*d
    return np.array([[aa+bb-cc-dd, 2*(bc+ad), 2*(bd-ac)],
                     [2*(bc-ad), aa+cc-bb-dd, 2*(cd+ab)],
                     [2*(bd+ac), 2*(cd-ab), aa+dd-bb-cc]])
```
