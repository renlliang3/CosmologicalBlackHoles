```python


#==================================================================================================
#Phase diagram of a the gas particles for a given snapshot and for given simulation
#==================================================================================================
def ScatterDiagram( prop1, prop2, prop3, 
                    label1, label2, label3, 
                    Xlim=None, Ylim=None, Clim=None ):
   
    #User defined ranges
    if Xlim == None:
        Xlim = [prop1.min(), prop1.max()]
    if Ylim == None:
        Ylim = [prop2.min(), prop2.max()]
    if Clim == None:
        Clim = [prop3.min(), prop3.max()]
    
    prop1_hist = 5
    prop2_hist = 5
    #Bins of 1D histograms
    bins = 28*2.
    
    #Formating plot....................................................
    #start with a rectangular Figure
    fig = plt.figure(1, figsize=(8,8))
    #no labels
    nullfmt = NullFormatter()
    #definitions for the axes
    left, width = 0.1, 0.65
    bottom_v, height = 0.1, 0.65
    bottom_h = left_h = left+width+0.02
    
    rect_hist2D = [left, bottom_v, width, height]
    rect_histx = [left, bottom_h, 1.24*width, 0.2]
    rect_histy = [left_h, bottom_v, 0.2, height]

    axHist2D = fig.add_axes(rect_hist2D)
    axHistx = fig.add_axes(rect_histx)
    axHisty = fig.add_axes(rect_histy)
    
    #no labels
    axHistx.xaxis.set_major_formatter(nullfmt)
    axHisty.yaxis.set_major_formatter(nullfmt)
     
    #Scatter2D.......................................................
    map2d = axHist2D.scatter( prop1, prop2, c = prop3, lw = 0, cmap = 'jet' )
    #Create the colorbar
    axc, kw = matplotlib.colorbar.make_axes( axHistx,\
    orientation = "vertical", shrink=1., pad=.04, aspect=10, anchor=(.5,1.5) )
    cb = matplotlib.colorbar.Colorbar( axc, map2d, orientation = "vertical" )
    cb.set_label( label3,  labelpad=-70, fontsize=12 )
    ticks = np.linspace( Clim[0],Clim[1],9 )
    cb.set_ticks( ticks )
    cb.set_ticklabels( ['{0:3.1f}'.format(t) for t in ticks] )
    
    #Histogram X (Prop1)
    histx = np.histogram( prop1, bins=bins, normed=True, range=Xlim )
    axHistx.bar( histx[1][:-1], histx[0], width = (Xlim[1]-Xlim[0])/bins,\
    linewidth=2.0, color="gray" )
    #Histogram Y (Prop2)
    histy = np.histogram( prop2, bins=bins, normed=True, range=Ylim )
    axHisty.barh( histy[1][:-1], histy[0], height = (Ylim[1]-Ylim[0])/bins,\
    linewidth=2.0, color="gray" )
    
    #Plot format.......................................................
    #Upper histogram
    axHistx.set_xlim( Xlim )
    axHistx.set_xticks( np.linspace( Xlim[0],Xlim[1],bins/4.-1 ) )
    yticks = np.linspace( 0, np.max(histx[0]), prop1_hist+1 )
    axHistx.set_yticks( yticks )
    axHistx.set_yticklabels( ['{0:1.2f}'.format(yt) for yt in yticks] )
    axHistx.grid( color='black', linestyle='--', linewidth=1., alpha=0.3 )
    axHistx.set_ylabel( "Normed distribution" )
    
    #Right histogram
    axHisty.set_ylim( Ylim )
    axHisty.set_yticks( np.linspace( Ylim[0],Ylim[1],bins/4.-1 ) )
    xticks = np.linspace( 0, np.max(histy[0]), prop2_hist+1 )
    axHisty.set_xticks( xticks )
    axHisty.set_xticklabels( ['{0:1.2f}'.format(xt) for xt in xticks], rotation=-90 )
    axHisty.grid( color='black', linestyle='--', linewidth=1., alpha=0.3 )
    axHisty.set_xlabel( "Normed distribution" )
    
    #2D Scatter
    axHist2D.set_xticks( xticks, ['{0:1.2f}'.format(xt) for xt in xticks] )
    yticks = np.linspace( Ylim[0],Ylim[1],bins/4.-1 )
    axHist2D.set_yticks( yticks, ['{0:1.2f}'.format(yt) for yt in yticks] )
    axHist2D.set_xlabel( label1, fontsize=14 )
    axHist2D.set_ylabel( label2, fontsize=14 )
    axHist2D.legend( loc='upper right', fancybox=True, shadow=True, ncol = 1, prop={'size':10} )
    axHist2D.grid(1)
    axHist2D.set_xlim( Xlim )
    axHist2D.set_ylim( Ylim )
    return axHist2D
```

## BHs in subfind subhalos

[Contents](#Contents)

```python
#Loading subfind subhalos
sub = arepo.Subfind('%s/%s/groups_015'%(DataFolder, Simulation), snapshot=15, combineFiles=True )
```

```python
plt.figure( figsize = (10,6) )
plt.title( "SBHs per Subhalo" )
plt.hist( sub.SubhaloLenType[:,-1], range=(-0.5,3.5), bins = 4, color='gray',
          weights = 100*np.ones(sub.Nsubgroups_Total)/sub.Nsubgroups_Total)
plt.xlabel( 'Number of BHs' )
plt.ylabel( 'Percentage of Subhalos' )
plt.grid()
plt.xticks( (0,1,2) )[-1]
```

```python
#Black Hole properties
d_BH_MP = []         #BH-Minimum potential distance
BH_mass = []         #BH mass
BH_tmass = []         #BH total mass
BH_pos = []          #BH position
BH_vel = []          #BH velocity
BH_pot = []          #BH potential
Sub_mass = []        #Host Subhalo mass
MP_pos = []          #Minimum potential position

i_c = 0
for i in xrange(sub.Nsubgroups_Total):
    #Detecting how many BHs the current subhalo has
    if( sub.SubhaloBHMass[i] == 0 ): continue
    #Filtering snapshot to recover most massive BH of current subhalo
    filter = arepo.filter.Halo( sub, subhalo=i )
    sn = arepo.Snapshot( '%s/%s/snapdir_015'%(DataFolder, Simulation), snapshot=15, combineFiles=True, parttype=[5], filter=filter )
    #Storing properties
    BH_pos.append( sn.part5.pos[0] )
    BH_vel.append( sn.part5.vel[0] )
    BH_mass.append( sn.part5.BH_Mass[0] )
    BH_tmass.append( sn.part5.Masses[0] )
    BH_pot.append( sn.part5.Potential[0] )
    Sub_mass.append( sub.SubhaloMass[i] )    
    MP_pos.append( sub.SubhaloPos[i] )
    
    #Taking position of minimum of the potential for current subhalo
    d_BH_MP.append( np.linalg.norm( BH_pos[i_c] - MP_pos[i_c] ) )
    
    i_c += 1
    
d_BH_MP = np.array(d_BH_MP)
BH_mass = np.array(BH_mass)
BH_tmass = np.array(BH_tmass)
BH_pos = np.array(BH_pos)
BH_vel = np.array(BH_vel)
BH_pot = np.array(BH_pot)
Sub_mass = np.array(Sub_mass)
MP_pos = np.array(MP_pos)
```
