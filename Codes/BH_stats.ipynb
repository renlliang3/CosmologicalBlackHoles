```python
#==================================================================================================
#Importing all libraries
#==================================================================================================
%pylab inline
from IPython.display import HTML
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
# JSAnimation import available at https://github.com/jakevdp/JSAnimation
from JSAnimation import IPython_display
from matplotlib import animation
import matplotlib.image as mpimg
import numpy as np
import subprocess
import os
import h5py
import arepo
```

```python
#Data folder
DataFolder = '/home/bustamsn/PhD/Data/Illustris/'
#Simulation
Simulation = 'L25n256'
```

# BHs in subfind subhalos

```python
#Loading subfind subhalos
sub = arepo.Subfind('%s/%s/groups_015'%(DataFolder, Simulation), snapshot=15, combineFiles=True )
```

```python
plt.figure( figsize = (10,6) )
plt.title( "Number of BHs per Subhalo" )
plt.hist( sub.SubhaloLenType[:,-1], range=(-0.5,3.5), bins = 4 )
plt.xlabel( 'Number of BHs' )
plt.ylabel( 'Number of Subhalos' )
plt.xticks( (0,1,2) )[-1]
```

```python
#Calculating distances
d_BH_MP = []
BH_mass = []
Sub_mass = []
for i in xrange(sub.Nsubgroups_Total):
    #Detecting how many BHs the current subhalo has
    if( sub.SubhaloBHMass[i] == 0 ): continue
    #Filtering snapshot to recover most massive BH of current subhalo
    filter = arepo.filter.Halo( sub, subhalo=i )
    sn = arepo.Snapshot( '%s/%s/snapdir_015'%(DataFolder, Simulation), snapshot=15, combineFiles=True, parttype=[5], filter=filter )
    BH_pos = sn.part5.pos[0]
    #Taking position of minimum of the potential for current subhalo
    MP_pos = sub.SubhaloPos[i]
    #Storing
    d_BH_MP.append( np.linalg.norm( BH_pos - MP_pos ) )
    BH_mass.append( sn.part5.BH_Mass[0] )
    Sub_mass.append( sub.SubhaloMass[i] )
```

## Distance of the main BH to the minimum of the potential

```python
plt.figure( figsize = (18,6) )

#Linear scale
plt.subplot(1,2,1)
plt.hist(d_BH_MP, bins=50 )[-1]
plt.grid()
plt.ylabel( 'Number of Subhalos', fontsize=14 )
plt.xlabel( '$d\ [kpc]$', fontsize=16 )

#Log scale
plt.subplot(1,2,2)
plt.hist(np.log10(d_BH_MP), bins=50 )[-1]
plt.grid()
plt.ylabel( 'Number of Subhalos', fontsize=14 )
plt.xlabel( '$\log_{10}( d/[kpc] )$', fontsize=16 )
```

## Relation between total mass of the subhalo and its most massive BH

```python
plt.figure( figsize=(10,6) )
plt.loglog( Sub_mass, BH_mass, '.' )
plt.xlabel( 'Subhalo mass $[1 \\times 10^{10} M_{\odot}]$', fontsize=16 )
plt.ylabel( 'BH mass $[1 \\times 10^{10} M_{\odot}]$', fontsize=16 )
plt.grid()
```
