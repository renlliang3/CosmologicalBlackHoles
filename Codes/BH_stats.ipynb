## Spin Parameter Distribution

### Fix BH + spin evolution (no recoil)

```python
plt.figure( figsize=(10,10) )
plt.subplots_adjust( left=0.06, right=0.96, top=0.98, wspace=0.00, hspace=0.0, bottom=0.09 )
for i, snaps in zip(xrange(1,10), xrange(7,16)):
    plt.subplot( 3,3,i )
    sn = arepo.Snapshot( '%s/%s/output/snapdir_000'%(DataFolder, Simulation3), snapshot=snaps, combineFiles=True, parttype=[5] )
    plt.plot( np.log10(sn.BH_Mass[sn.BH_SpinModel==1]*1e10), sn.BH_SpinParameter[sn.BH_SpinModel==1], '.', color='red', label='Chaotic' )
    plt.plot( np.log10(sn.BH_Mass[sn.BH_SpinModel==0]*1e10), sn.BH_SpinParameter[sn.BH_SpinModel==0], '.', color='blue', label='Prolonged' )

    plt.ylim( (-0.2, 1.1) )
    plt.xlim( (5, 10) )
    plt.text(9, 1, 'z=%1.2f'%(sn.Redshift))
    if i<7:
        plt.xticks([])
    else:
        plt.xticks([6,9])
    if i==1 or i==4 or i==7:
        plt.yticks([0,1])
    else:
        plt.yticks([])
    if i==8:
        plt.xlabel( '$\log_{10}(M_{BH}/M_{\odot} )$', fontsize = 16 )
    if i==4:
        plt.ylabel( 'Spin parameter', fontsize = 16 )
    if i==9:
        plt.legend( loc='upper left', fancybox=True )
```

## Mass Accretion Rate Distribution

### Fix BH + spin evolution (no recoil)

```python
plt.figure( figsize=(10,10) )
plt.subplots_adjust( left=0.06, right=0.96, top=0.98, wspace=0.00, hspace=0.0, bottom=0.09 )
for i, snaps in zip(xrange(1,10), xrange(7,16)):
    plt.subplot( 3,3,i )
    sn = arepo.Snapshot( '%s/%s/output/snapdir_000'%(DataFolder, Simulation3), snapshot=snaps, combineFiles=True, parttype=[5] )
    plt.plot( np.log10(sn.BH_Mass[sn.BH_SpinModel==1]*1e10), np.log10(sn.BH_MdotBondi[sn.BH_SpinModel==1]/sn.BH_MdotEddington[sn.BH_SpinModel==1]), '.', color='red', label='Chaotic' )
    plt.plot( np.log10(sn.BH_Mass[sn.BH_SpinModel==0]*1e10), np.log10(sn.BH_MdotBondi[sn.BH_SpinModel==0]/sn.BH_MdotEddington[sn.BH_SpinModel==0]), '.', color='blue', label='Prolonged' )

    plt.ylim( (-7, 0) )
    plt.xlim( (5, 10) )
    plt.hlines( -3, 5, 10, linestyles='--', linewidths = 0.5, colors='black' )
    plt.text(9, 1, 'z=%1.2f'%(sn.Redshift))
    if i<7:
        plt.xticks([])
    else:
        plt.xticks([6,9])
    if i==1 or i==4 or i==7:
        plt.yticks([-6,-1])
    else:
        plt.yticks([])
    if i==8:
        plt.xlabel( '$\log_{10}(M_{BH}/M_{\odot} )$', fontsize = 16 )
    if i==4:
        plt.ylabel( '$\log_{10}( \dot m )$', fontsize = 16 )
    if i==9:
        plt.legend( loc='upper left', fancybox=True )
```

## Spin Alignment

```python
#Loading all particles at z=0
sn_DM = arepo.Snapshot( '%s/%s/output/snapdir_000'%(DataFolder, Simulation4), snapshot=15, combineFiles=True,parttype=[0] )
sn_BH = arepo.Snapshot( '%s/%s/output/snapdir_000'%(DataFolder, Simulation4), snapshot=15, combineFiles=True,parttype=[5] )
```

```python
plt.figure( figsize = (15,15) )
plt.subplots_adjust( left=0.06, right=0.96, top=0.98, wspace=0.00, hspace=0.0, bottom=0.09 )
for i in xrange(4):
    plt.subplot( 2,2,i+1 )
    #Slices
    Lmin = 5000*i-2500
    Lmax = 5000*i+2500
    
    mask_slice_gas = (sn_DM.Coordinates[:,2]<Lmax)*(sn_DM.Coordinates[:,2]>=Lmin)
    mask_slice_BH = (sn_BH.Coordinates[:,2]<Lmax)*(sn_BH.Coordinates[:,2]>=Lmin)
    
    if Lmin < 0:
        Lmin = 25000-2500
        mask_slice_gas = ((sn_DM.Coordinates[:,2]<Lmax).astype(int)+(sn_DM.Coordinates[:,2]>=Lmin).astype(int))>=1
        mask_slice_BH = ((sn_BH.Coordinates[:,2]<Lmax).astype(int)+(sn_BH.Coordinates[:,2]>=Lmin).astype(int))>=1
    #dL = 15000
    #dLz = 1000
    #G = 25
    #mask_slice_gas = (abs(sn_DM.Coordinates[:,0]-sub3.GroupCM[G,0])<dL)*(abs(sn_DM.Coordinates[:,1]-sub3.GroupCM[G,1])<dL)*(abs(sn_DM.Coordinates[:,2]-sub3.GroupCM[G,2])<dLz)
    #mask_slice_BH = (abs(sn_BH.Coordinates[:,0]-sub3.GroupCM[G,0])<dL)*(abs(sn_BH.Coordinates[:,1]-sub3.GroupCM[G,1])<dL)*(abs(sn_BH.Coordinates[:,2]-sub3.GroupCM[G,2])<dLz)
    
    mask_BH_chaotic = sn_BH.BH_SpinModel[mask_slice_BH] == 1
    mask_BH_prolong = sn_BH.BH_SpinModel[mask_slice_BH] == 0
    
    plt.plot( sn_DM.Coordinates[mask_slice_gas,0], sn_DM.Coordinates[mask_slice_gas,1], '.', ms=0.05, color='gray' )
    
    plt.plot( sn_BH.Coordinates[mask_slice_BH,0][mask_BH_chaotic], sn_BH.Coordinates[mask_slice_BH,1][mask_BH_chaotic], '.', ms=10, color='red' )
    plt.plot( sn_BH.Coordinates[mask_slice_BH,0][mask_BH_prolong], sn_BH.Coordinates[mask_slice_BH,1][mask_BH_prolong], '.', ms=10, color='blue' )
    
    Q = plt.quiver( sn_BH.Coordinates[mask_slice_BH,0][mask_BH_chaotic], sn_BH.Coordinates[mask_slice_BH,1][mask_BH_chaotic], 
    sn_BH.BH_SpinOrientation[mask_slice_BH,0][mask_BH_chaotic], sn_BH.BH_SpinOrientation[mask_slice_BH,1][mask_BH_chaotic], 
    zorder = 100, width=0.002, color='red' )
    
    Q = plt.quiver( sn_BH.Coordinates[mask_slice_BH,0][mask_BH_prolong], sn_BH.Coordinates[mask_slice_BH,1][mask_BH_prolong], 
    sn_BH.BH_SpinOrientation[mask_slice_BH,0][mask_BH_prolong], sn_BH.BH_SpinOrientation[mask_slice_BH,1][mask_BH_prolong], 
    zorder = 100, width=0.002, color='blue' )
    plt.quiverkey(Q,0.04,0.01,1,'1')
    
    plt.xticks([])
    plt.yticks([])
    #plt.xlim( (sub3.GroupCM[G,0]-dL, sub3.GroupCM[G,0]+dL) )
    #plt.ylim( (sub3.GroupCM[G,1]-dL, sub3.GroupCM[G,1]+dL) )
```

```python
plt.figure( figsize=(10,5) )
mask_BH_chaotic = sn_BH.BH_SpinModel == 1
mask_BH_prolong = sn_BH.BH_SpinModel == 0

AngMomNorm = (sn_BH.BH_AngMomGasCells/np.linalg.norm(sn_BH.BH_AngMomGasCells,axis=1).reshape(-1,1))
angle = np.sum(sn_BH.BH_SpinOrientation*AngMomNorm,axis=1)
plt.hist(angle[mask_BH_chaotic], bins=50, alpha=0.5, normed=True, color='red', label='Chaotic')[-1]
plt.hist(angle[mask_BH_prolong], bins=50, alpha=0.5, normed=True, color='blue', label='Prolonged')[-1]
plt.legend( loc='upper left' )
```
