## BHs in subfind subhalos

[Contents](#Contents)

```python
#Loading subfind subhalos
sub1 = arepo.Subfind('%s/%s/output/groups_015'%(DataFolder, Simulation2), snapshot=15, combineFiles=True )
sub2 = arepo.Subfind('%s/%s/output/groups_015'%(DataFolder, Simulation1), snapshot=15, combineFiles=True )
```

```python
plt.figure( figsize = (10,10) )
plt.scatter( np.log10(sub1.SubhaloMassType[:,4]*1e10), np.log10(sub1.SubhaloBHMass*1e10), color='blue' )
plt.scatter( np.log10(sub2.SubhaloMassType[:,4]*1e10), np.log10(sub2.SubhaloBHMass*1e10), color='red' )
plt.xlim( (8.5, 12) )
plt.ylim( (4, 11) )
plt.grid()
```

```python
#Black Hole properties
d_BH_MP1 = []         #BH-Minimum potential distance
BH_mass1 = []         #BH mass
BH_tmass1 = []         #BH total mass
BH_pos1 = []          #BH position
BH_vel1 = []          #BH velocity
BH_pot1 = []          #BH potential
Sub_mass1 = []        #Host Subhalo mass
MP_pos1 = []          #Minimum potential position

i_c = 0
for i in xrange(sub1.Nsubgroups_Total):
    #Detecting how many BHs the current subhalo has
    if( sub1.SubhaloBHMass[i] == 0 ): continue
    #Filtering snapshot to recover most massive BH of current subhalo
    filter = arepo.filter.Halo( sub1, subhalo=i )
    sn = arepo.Snapshot( '%s/%s/output/snapdir_015'%(DataFolder, Simulation1), snapshot=15, combineFiles=True, parttype=[5], filter=filter )
    #Storing properties
    BH_pos1.append( sn.part5.pos[0] )
    BH_vel1.append( sn.part5.vel[0] )
    BH_mass1.append( sn.part5.BH_Mass[0] )
    BH_tmass1.append( sn.part5.Masses[0] )
    BH_pot1.append( sn.part5.Potential[0] )
    Sub_mass1.append( sub1.SubhaloMass[i] )    
    MP_pos1.append( sub1.SubhaloPos[i] )
    
    #Taking position of minimum of the potential for current subhalo
    d_BH_MP1.append( np.linalg.norm( BH_pos1[i_c] - MP_pos1[i_c] ) )
    
    i_c += 1
    
d_BH_MP1 = np.array(d_BH_MP1)
BH_mass1 = np.array(BH_mass1)
BH_tmass1 = np.array(BH_tmass1)
BH_pos1 = np.array(BH_pos1)
BH_vel1 = np.array(BH_vel1)
BH_pot1 = np.array(BH_pot1)
Sub_mass1 = np.array(Sub_mass1)
MP_pos1 = np.array(MP_pos1)
```

```python
plt.figure( figsize = (18,6) )

#Linear scale
plt.subplot(1,2,1)
plt.hist(d_BH_MP1, bins=50, color='gray' )[-1]
plt.grid()
plt.ylabel( 'Number of Subhalos', fontsize=14 )
plt.xlabel( '$d\ [kpc]$', fontsize=16 )

#Log scale
plt.subplot(1,2,2)
plt.hist(np.log10(d_BH_MP1), bins=50, color='gray' )[-1]
plt.grid()
plt.ylabel( 'Number of Subhalos', fontsize=14 )
plt.xlabel( '$\log_{10}( d/[kpc] )$', fontsize=16 )
```
