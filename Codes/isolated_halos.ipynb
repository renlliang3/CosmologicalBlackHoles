# Control Halo with no perturber

For this set of controlled experiments, we use the Hernquist profile given by:

$$  \rho = \frac{M}{2\pi}\frac{a}{r}\frac{1}{\left( r+a \right)^3} $$

$$ \Phi = - \frac{GM}{r+a} $$

$$ M(r) = M\frac{r^2}{(r+a)^2} $$

Where $M = 2.0 \times 10^{12} M_{\odot}$ is the total mass of the halo and $a = 40 \mbox{kpc}$ is the scale length for all the setups.

## Distribution 

An interesting property of Herquinst profiles is that the associated distributions are static over time.
For this setup we integrate $100000$ DM particles over a period of $t_{max} = 5 \mbox{Gyr}$. Plotting the particle distribution for mass profile for an initial, intermediate and final snapshots we got:

```python
N_part = 100000.
M_tot = 200.0
a_hern = 40.0
```

```python
plt.figure( figsize = (20,5) )
i = 1
for snap in [0, 25, 51]:
    plt.subplot(1,3,i)
    data = gadget.Simulation("%s/%s/output/snapshot_%03d.hdf5"%(DataFolder, 'ConNP', snap))
    plt.plot(data.part1.pos[:,0], data.part1.pos[:,1],'.')
    plt.title( 'Snapshot %d'%(snap) )
    plt.xlabel( 'X  [kpc]' )
    plt.ylabel( 'Y  [kpc]' )
    i += 1
```

```python
plt.figure( figsize = (20,5) )
i = 1
#Hernquist mass profile
lRh = np.linspace( 0, 4, 100 )
M_hern = M_tot*(10**lRh)**2/( 10**Rh + a_hern )**2

for snap in [0, 25, 51]:
    plt.subplot(1,3,i)
    data = gadget.Simulation("%s/%s/output/snapshot_%03d.hdf5"%(DataFolder, 'ConNP', snap))
    plt.hist(np.log10(data.r()), bins=100, cumulative=True, weights = M_tot*np.ones( N_part )/N_part, histtype='step', label='Numerical')[-1]
    plt.plot( Rh, M_hern, lw = 2, label = 'Hernquist profile' )
    plt.title( 'Snapshot %d'%(snap) )
    plt.xlabel( '$\log_{10}( r/kpc )$', fontsize = 15 )
    plt.ylabel( '$M$  [$10^{10}M_{\odot}$]', fontsize = 15 )
    plt.xlim( (0,4) )
    plt.grid()
    i += 1
plt.legend( loc='lower right' )
```

```python
plt.figure( figsize = (20,5) )
i = 1
#Hernquist mass profile
lRh = np.linspace( 0, 4, 100 )
M_hern = M_tot*(10**lRh)**2/( 10**Rh + a_hern )**2

for snap in [0, 25, 51]:
    plt.subplot(1,3,i)
    data = gadget.Simulation("%s/%s/output/snapshot_%03d.hdf5"%(DataFolder, 'ConNP', snap))
    V = np.ones( N_part )
    for iv in xrange(3):
        V += data.part1.vel[:,iv]**2
    V = V**0.5
    plt.hist( V, bins=100, histtype='step')[-1]
    plt.title( 'Snapshot %d'%(snap) )
    plt.xlabel( '$\log_{10}( r/kpc )$', fontsize = 15 )
    plt.ylabel( '$M$  [$10^{10}M_{\odot}$]', fontsize = 15 )
    plt.grid()
    i += 1
plt.legend( loc='lower right' )
```
